
# 📦 УРОК 03 — Змінні та типи даних
## Ключова ідея: Змінна — це ярлик, а не коробка

---

## 1. Змінні — посилання на об'єкти

```
a = 7      →  імʼя 'a' ─────────────► [📦 об'єкт: 7]
b = a      →  імʼя 'b' ─────────────► [📦 об'єкт: 7]   ← ТОЙ САМИЙ об'єкт!
a = 10     →  імʼя 'a' ──────────────────────────────► [📦 об'єкт: 10]
             імʼя 'b' ─────────────► [📦 об'єкт: 7]   ← b НЕ змінився!
```

---

## 2. Примітивні типи

| Тип | Приклад | Особливість |
|-----|---------|-------------|
| `int` | `42`, `-5` | Необмежена довжина в Python 3 |
| `float` | `3.14`, `-1.2` | IEEE 754, є похибки (0.1+0.2 ≠ 0.3) |
| `str` | `"text"`, `'hello'` | **Незмінний (immutable)!** |
| `bool` | `True`, `False` | Підклас `int`: True=1, False=0 |
| `None` | `None` | Singleton — відсутність значення |

---

## 3. Типізація Python: динамічна + строга

```python
x = 10        # тип int — не треба оголошувати
x = "hello"   # тепер str — Python дозволяє

"spam" + 5    # ❌ TypeError — Python СУВОРО забороняє мішати типи!
```

> **Duck Typing:** Python не питає хто ти — він питає що ти вмієш.
> Якщо об'єкт підтримує `+` — можна додавати.

---

## 4. Перетворення типів (явне!)

```python
int("42")      # → 42
int(3.9)       # → 3  (УСІЧЕННЯ, не округлення!)
float(5)       # → 5.0
str(100)       # → "100"
```

---

## 5. Рядки — послідовність символів з кодами

```python
word = "Python"
word[0]        # 'P'   — перший символ
word[-1]       # 'n'   — останній
ord('A')       # 65    — символ → Unicode код
chr(65)        # 'A'   — код → символ
```

> **Рефлексія:** Текст — це числа. Шифрування, NLP, AI — все будується на цьому.

````python
# Демонстрація: змінна — це ярлик
a = 7
b = a
a = 10

print(f"a = {a}, b = {b}")  # b залишився 7!
print()

# Динамічна, але СТРОГА типізація
print("int + int:", 5 + 3)         # ✅
print("str + str:", "Py" + "thon") # ✅
try:
    print("str + int:", "age: " + 25)
except TypeError as e:
    print(f"str + int → TypeError: {e}")  # ❌

print()

# int() = усічення, НЕ округлення!
print(f"int(3.9) = {int(3.9)}   ← відкидає дробову частину")
print(f"round(3.9) = {round(3.9)} ← математичне округлення")

print()

# Пріоритет операцій
print(f"1 + 2 * 3 = {1 + 2 * 3}   (спочатку *, потім +)")
print(f"(1 + 2) * 3 = {(1 + 2) * 3} (дужки змінюють порядок)")
````

# 🔀 УРОК 04 — Булева логіка та потік виконання
## Ключова ідея: Програма приймає рішення через умови

---

## 1. True і False — що Python вважає брехнею

**Завжди False (Falsy):**

| Значення | Тип |
|----------|-----|
| `0` | int |
| `0.0` | float |
| `""` | str |
| `[]` | list |
| `{}` | dict |
| `None` | NoneType |

> **Все інше — True.** Навіть `"False"` (рядок з текстом False) — це True!

---

## 2. Логічні оператори та пріоритет

```
not  →  and  →  or   (від вищого до нижчого пріоритету)
```

**Short-circuiting:**
```python
True or f()    # f() НЕ викликається — результат і так True
False and f()  # f() НЕ викликається — результат і так False

# Практично: безпечне ділення
if y != 0 and x / y > 1:   # ділення не відбудеться якщо y == 0
    ...
```

---

## 3. if / elif / else — порядок має значення!

```python
# ❌ Неправильний порядок — elif ніколи не спрацює
if score >= 60:    # True для score=85 → входимо сюди і виходимо
    print("Склав")
elif score >= 80:  # Ніколи не перевіряється!
    print("B")

# ✅ Правильно: від конкретного до загального

elif score >= 80: print("B")

else:             print("Не склав")
```

## 4. while — повторення до події

```python

```

---

## 5. `__name__ == "__main__"` — два режими файлу

| Режим | `__name__` | Блок `if` виконується? |
|-------|-----------|------------------------|
| `python script.py` | `"__main__"` | ✅ так |
| `import script` | `"script"` | ❌ ні |

> **Правило:** Завжди захищай точку входу через `if __name__ == "__main__":`

````python
# Truthy / Falsy
falsy_values = [0, 0.0, "", [], {}, None]
print("Falsy значення:")
for v in falsy_values:
    print(f"  bool({v!r:8}) = {bool(v)}")

print()

# Пріоритет логічних операторів
x, y = True, False
print(f"not x and y or x = {not x and y or x}")  # ((not x) and y) or x
print("  Читається як: ((not True) and False) or True = True")

print()

# Порядок умов
for score in [95, 82, 65, 40]:
    if score >= 90:   grade = "A"
    elif score >= 80: grade = "B"
    elif score >= 60: grade = "Склав"
    else:             grade = "Не склав"
    print(f"  score={score} → {grade}")

print()

# break та continue
print("continue (пропускаємо 3, break на 5):")
for num in range(1, 7):
    if num == 3: continue
    if num == 5: break
    print(num, end=" ")
print()
````

# 📦 УРОК 05 — Модулі, імпорти та CLI
## Ключова ідея: Будь-який .py файл — це модуль

---

## 1. Що відбувається при `import`?

```
1. SEARCH   — шукає файл за sys.path
2. COMPILE  — компілює у байткод (.pyc)
3. EXECUTE  — виконує весь top-level код
4. CACHE    — зберігає у sys.modules (повторний import = кеш)
```

> ⚠️ **Top-level код виконується при import!** Тому `print()` і side effects
> на рівні файлу — погана практика. На верхньому рівні — тільки `def`, `class`, константи.

---

## 2. Форми імпорту

| Синтаксис | Використання |
|-----------|--------------|
| `import math` | `math.sqrt(4)` — весь модуль |
| `from math import sqrt` | `sqrt(4)` — конкретна функція |
| `import numpy as np` | `np.array(...)` — псевдонім |
| `from math import *` | ⚠️ забруднює namespace — уникати! |

---

## 3. sys.path та Shadowing

```
Порядок пошуку:
  1. Поточна директорія  ← найвищий пріоритет!
  2. PYTHONPATH
  3. Стандартна бібліотека
  4. site-packages (pip install)
```

> ⚠️ Якщо назвеш файл `random.py` — Python знайде його ПЕРШИМ і не зможе
> завантажити стандартний `random`! Ніколи не давай файлам імена стандартних модулів.

---

## 4. sys.argv — аргументи CLI

```python
# Запуск: python script.py Alice 42
import sys
sys.argv[0]   # 'script.py'  ← ім'я скрипта
sys.argv[1]   # 'Alice'      ← завжди STR!
```

# 🗃️ УРОК 06 — Списки, кортежі та множини
## Ключова ідея: Три різних способи зберігати колекцію

---

## Порівняння контейнерів

| | `list []` | `tuple ()` | `set {}` |
|--|-----------|------------|----------|
| **Порядок** | ✅ Впорядкований | ✅ Впорядкований | ❌ Невпорядкований |
| **Змінність** | ✅ Змінний | ❌ Незмінний | ✅ Змінний |
| **Дублікати** | ✅ Дозволені | ✅ Дозволені | ❌ Тільки унікальні |
| **Індекс** | ✅ `lst[0]` | ✅ `tpl[0]` | ❌ Немає |
| **Хешується** | ❌ | ✅ | ❌ (frozenset ✅) |

---

## Коли що використовувати?

- **list** → дані динамічні, порядок важливий (кошик, черга задач)
- **tuple** → фіксовані дані, захист від змін (координати, налаштування)
- **set** → потрібна унікальність або швидкий пошук O(1)

---

## Ключові операції

### Списки: індексація та зрізи
```python
lst = ['a', 'b', 'c', 'd', 'e']
#       0    1    2    3    4
#      -5   -4   -3   -2   -1
lst[1:4]    # ['b', 'c', 'd']
lst[::-1]   # ['e', 'd', 'c', 'b', 'a']  ← реверс
lst[::2]    # ['a', 'c', 'e']  ← кожен другий
```

### Кортежі: розпакування
```python
a, b, c = (1, 2, 3)         # розпакування
first, *rest = (1, 2, 3, 4) # зі зіркою
a, b = b, a                  # обмін без temp!
```

### Множини: теорія множин
```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
A | B   # {1,2,3,4,5,6}  — об'єднання
A & B   # {3,4}           — перетин
A - B   # {1,2}           — різниця
A ^ B   # {1,2,5,6}       — симетрична різниця
```

---

## Типові помилки

```python
{}      # ← СЛОВНИК, не множина! Порожня множина = set()
(42)    # ← int! Кортеж з одним елементом = (42,)  ← кома!
```

> ⚠️ Ніколи не видаляй елементи зі списку **під час ітерації** по ньому.
> Ітеруй по копії: `for item in my_list.copy():`

````python
# Порівняння контейнерів
raw = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
print(f"list:  {list(raw)}   (всі елементи, порядок збережено)")
print(f"tuple: {tuple(raw)}  (незмінний)")
print(f"set:   {set(raw)}    (тільки унікальні, порядок довільний)")
print()

# Зрізи
fruits = ['яблуко', 'банан', 'апельсин', 'ківі', 'манго']
print(f"fruits[1:4]  = {fruits[1:4]}")
print(f"fruits[::-1] = {fruits[::-1]}")
print()

# Кортеж: розпакування
point = (10, 20, 30)
x, y, z = point
first, *rest = (1, 2, 3, 4, 5)
print(f"x={x}, y={y}, z={z}")
print(f"first={first}, rest={rest}")
print()

# Множини: операції
A, B = {1, 2, 3, 4}, {3, 4, 5, 6}
print(f"A | B = {A | B}")
print(f"A & B = {A & B}")
print(f"A - B = {A - B}")
print()

# Швидкість пошуку
import time
big = list(range(1_000_000))
big_set = set(big)
target = 999_999

t0 = time.time(); _ = (target in big); t_list = time.time() - t0
t0 = time.time(); _ = (target in big_set); t_set = time.time() - t0
print(f"list пошук: {t_list*1000:.4f} мс")
print(f"set  пошук: {t_set*1000:.4f} мс")
print(f"set швидший у {t_list/max(t_set,1e-9):.0f}x разів  (O(1) vs O(n))")
````

# 🔄 УРОК 07 — Цикли, словники та Comprehensions
## Ключова ідея: Чотири патерни обробки даних

---

## Як мислить програміст?

```
1. ITERATION  → Як перебрати дані?          (for, while)
2. MAPPING    → Як трансформувати кожен?    ([f(x) for x in ...])
3. COUNTING   → Як підрахувати частоти?     (Counter, defaultdict(int))
4. AGGREGATION→ Як зібрати у одне значення? (sum, max, reduce)
```

---

## for vs while

| | `for` | `while` |
|--|-------|----------|
| Управляється | Колекцією | Умовою |
| Коли | Відома колекція | До настання події |
| Безпечний? | ✅ Python сам слідкує | ⚠️ Ти сам маєш змінювати умову |

---

## Ітерація по словниках

```python
d = {'Alice': 95, 'Bob': 82}

for key in d:              # → 'Alice', 'Bob'
for key in d.keys():       # → 'Alice', 'Bob'
for val in d.values():     # → 95, 82
for k, v in d.items():     # → ('Alice', 95), ('Bob', 82)

# ❌ Помилка: for k, v in d  ← ValueError! Рядок не можна розпакувати у 2
```

---

## Безпечна робота зі словниками

## Comprehensions — «приплюснутий for»

```python
# Формула:
[ вираз   for елемент in колекція   if умова ]
#  MAP         ITERATE                FILTER

# List comprehension

# Dict comprehension

# Set comprehension
```

> ⚠️ Comprehension — тільки коли мета **побудувати колекцію**.
> Якщо мета — **виконати дію** (print, записати файл) → звичайний `for`.

````python
from collections import Counter, defaultdict

# --- ITERATION: enumerate ---
fruits = ['яблуко', 'банан', 'апельсин']
print("enumerate():")
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}. {fruit}")
print()

# --- MAPPING: трансформація ---
prices_usd = [10.5, 25.0, 3.99]
prices_uah = [round(p * 41.5, 2) for p in prices_usd]
print(f"USD: {prices_usd}")
print(f"UAH: {prices_uah}")
print()

# --- COUNTING: Counter ---
text = "banana"
freq = Counter(text)
print(f"Counter({text!r}) = {freq}")
print(f"most_common(2) = {freq.most_common(2)}")
print()

# --- AGGREGATION: один прохід, кілька метрик ---
scores = [85, 92, 78, 95, 88, 60, 73]
total = count_above_80 = 0
maximum = scores[0]

for s in scores:
    total += s
    if s > maximum: maximum = s
    if s >= 80: count_above_80 += 1

print(f"scores = {scores}")
print(f"Середнє: {total/len(scores):.1f}, Макс: {maximum}, Вище 80: {count_above_80}")
print()

# --- Grouping: defaultdict ---
purchases = [('Food', 15), ('Transport', 2), ('Food', 30), ('Books', 10)]
grouped = defaultdict(list)
for cat, amount in purchases:
    grouped[cat].append(amount)
for cat, amounts in grouped.items():
    print(f"  {cat}: {amounts} → сума {sum(amounts)}")
````

### Повна сигнатура — порядок обов'язковий!

```python
#            звичайні   довільні   з дефолтом   довільні іменовані
def full_fn(a, b,      *args,     c=10,        **kwargs):
    print(a, b, args, c, kwargs)

full_fn(1, 2, 3, 4, c=99, x='hello', y=True)
# a=1, b=2, args=(3, 4), c=99, kwargs={'x': 'hello', 'y': True}
```

| Параметр | Тип всередині | Що «ловить» |
|----------|--------------|-------------|
| `*args` | `tuple` | зайві **позиційні** аргументи |
| `**kwargs` | `dict` | зайві **іменовані** аргументи |

> ⚠️ Порядок: звичайні → `*args` → з дефолтом → `**kwargs`  
> `*args` і `**kwargs` — лише конвенція. Можна назвати `*values`, `**options`, але `args`/`kwargs` — стандарт.

````python
# return vs print
def bad_add(a, b): print(a + b)
def good_add(a, b): return a + b

bad_result = bad_add(2, 3)
print(f"bad_add result: {bad_result!r}  ← None!")
good_result = good_add(2, 3)
print(f"good_add result: {good_result}  × 10 = {good_result * 10}")
print()

# Чиста vs нечиста
def apply_discount_pure(prices, d):
    return [p * (1 - d) for p in prices]  # новий список

orig = [100.0, 200.0, 300.0]
result_pure = apply_discount_pure(orig, 0.10)
print(f"Чиста → результат: {result_pure}")
print(f"Чиста → оригінал:  {orig}  ← незмінений")
print()

# --- *args: довільна кількість позиційних ---
def sales_stats(*amounts):      # amounts — це tuple
    return {
        'count': len(amounts),
        'total': sum(amounts),
        'avg':   round(sum(amounts) / len(amounts), 2),
        'max':   max(amounts),
    }

print("=== *args ===")
stats = sales_stats(150, 300, 75, 200, 450)
for k, v in stats.items():
    print(f"  {k}: {v}")
print(f"  type(amounts) всередині → tuple")
print()

# --- **kwargs: довільна кількість іменованих ---
def create_user(**kwargs):      # kwargs — це dict
    defaults = {'role': 'student', 'active': True, 'score': 0}
    profile = {**defaults, **kwargs}   # kwargs перезаписує defaults
    return profile

print("=== **kwargs ===")
user1 = create_user(name='Alice', score=95)
user2 = create_user(name='Bob', role='admin', active=False)
print(f"  user1: {user1}")
print(f"  user2: {user2}")
print()

# --- *args + **kwargs разом ---
def log_event(event_type, *messages, level='INFO', **meta):
    print(f"[{level}] {event_type}")
    for msg in messages:
        print(f"  → {msg}")
    if meta:
        print(f"  meta: {meta}")

print("=== *args + **kwargs разом ===")
log_event('LOGIN', 'User Alice connected', level='INFO', ip='127.0.0.1', port=8080)
print()
log_event('ERROR', 'DB timeout', 'Retrying...', level='CRITICAL', retry=3)
print()

# --- Розпакування при виклику: * та ** ---
print("=== розпакування при виклику ===")
nums = [10, 20, 30]
opts = {'c': 99, 'x': 'hello'}

def demo(a, b, c=0, **kwargs):
    print(f"  a={a}, b={b}, c={c}, kwargs={kwargs}")

demo(*nums[:2], **opts)      # *розпаковує list, **розпаковує dict
````

## EAFP vs LBYL

| | LBYL (перевіряй) | EAFP (пробуй) |
|--|-----------------|---------------|
| Стиль | C, Java | **Python** |
| Код | Багато `if` перевірок | `try/except` |
| Проблема | Race conditions | Треба знати типи |

> **Python явно рекомендує EAFP** — «Легше вибачитись, ніж питати дозволу»

````python
# Демонстрація: поширення помилок по стеку
def function_c():
    return 10 / 0              # ZeroDivisionError тут

def function_b():
    return function_c()        # не перехоплює — передає вгору

def function_a():
    try:
        function_b()           # перехоплює з глибини!
    except ZeroDivisionError:
        print("A: Перехопив ZeroDivisionError з глибини стека")

function_a()
print("Програма продовжує роботу ✅")
print()

# try / except / else / finally
test_inputs = [("42", "10"), ("abc", "5"), ("10", "0")]

for a_str, b_str in test_inputs:
    try:
        a = int(a_str)
        b = int(b_str)
        result = a / b
    except ValueError:
        print(f"  {a_str}/{b_str} → ValueError: не число")
    except ZeroDivisionError:
        print(f"  {a_str}/{b_str} → ZeroDivisionError: ділення на 0")
    else:
        print(f"  {a_str}/{b_str} = {result:.2f} ✅")
    finally:
        pass  # тут закривали б ресурси

print()

# raise: захисне програмування
def validate_age(age):
    if not isinstance(age, int):
        raise TypeError("Вік має бути цілим числом")
    if age < 0 or age > 150:
        raise ValueError(f"Нереальний вік: {age}")
    return age

for test in [25, -5, "двадцять", 200]:
    try:
        print(f"  validate_age({test!r}) = {validate_age(test)} ✅")
    except (TypeError, ValueError) as e:
        print(f"  validate_age({test!r}) → {type(e).__name__}: {e}")
````

````python
import json
import tempfile, os

# Context Manager + запис
data = {
    "name": "Іван Петренко",
    "age": 28,
    "skills": ["Python", "SQL", "Git"],
    "active": True,
    "score": None
}

# Запис у тимчасовий файл
fname = tempfile.mktemp(suffix='.json')

with open(fname, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("Вміст JSON файлу:")
with open(fname, "r", encoding="utf-8") as f:
    content = f.read()
    print(content)

# Читання назад
with open(fname, "r", encoding="utf-8") as f:
    loaded = json.load(f)

print(f"Тип після json.load(): {type(loaded).__name__}  ← dict!")
print(f"name: {loaded['name']}, skills: {loaded['skills']}")
print()

# Обмеження: datetime
from datetime import datetime
try:
    json.dumps({"time": datetime.now()})
except TypeError as e:
    print(f"⚠️ datetime → TypeError: {e}")
    print(f"   Рішення: datetime.now().isoformat() → '{datetime.now().isoformat()}'")

os.unlink(fname)
print()

# Режим 'w' деструктивний!
print("⚠️ Режим 'w' СТИРАЄ попередній вміст!")
print("   Для дописування — режим 'a' (append)")
````

## Python-програма як Data Pipeline

```
📄 Файл / API / Ввід
      ↓  відкриваємо + читаємо (File I/O, json.load)
🔍 Фільтрація (Predicate)
      ↓  [x for x in data if is_valid(x)]
🔄 Трансформація (Transformer)
      ↓  [transform(x) for x in filtered]
📊 Агрегація (Reducer)
      ↓  sum(), Counter(), defaultdict()
💾 Збереження (File I/O, json.dump)
      ↓
📤 Відповідь / Звіт / API Response
```

````python
# === ПОВНИЙ PIPELINE — все, що вивчили у Модулі 1 ===
import json
from datetime import datetime
from collections import Counter, defaultdict

# Вхідні дані (імітація CSV або API відповіді)
raw_data = [
    "Alice,92,Food,150",
    "Bob,ERROR,Transport,30",     # пошкоджений рядок
    "Alice,85,Books,80",
    "Charlie,78,Food,200",
    "",                           # порожній рядок
    "Bob,91,Food,120",
    "Charlie,65,Transport,45",
    "Alice,88,Books,90",
]

print(f"Вхідних рядків: {len(raw_data)}")
print()

# --- Step 1: Filter (Predicate) ---
def is_valid_row(row):
    parts = row.split(',')
    return len(parts) == 4 and parts[1].strip().isdigit()

valid = [r for r in raw_data if is_valid_row(r)]
print(f"Step 1 Filter: {len(valid)}/{len(raw_data)} валідних рядків")

# --- Step 2: Map (Transformer) ---
def parse_row(row):
    name, score, category, amount = row.split(',')
    score_int = int(score)
    return {
        'name': name,
        'score': score_int,
        'grade': 'A' if score_int >= 90 else 'B' if score_int >= 80 else 'C',
        'category': category,
        'amount': int(amount),
    }

records = [parse_row(r) for r in valid]
print(f"Step 2 Map: {len(records)} записів")

# --- Step 3: Group + Reduce ---
by_name = defaultdict(list)
for rec in records:
    by_name[rec['name']].append(rec)

print()
print("=" * 50)
print("ЗВІТ СТУДЕНТІВ")
print("=" * 50)
print(f"{'Студент':<12} {'Середня':>8} {'Оцінка':>8} {'Витрати':>10}")
print("-" * 42)

summary = []
for name, recs in sorted(by_name.items()):
    avg_score = sum(r['score'] for r in recs) / len(recs)
    grade = 'A' if avg_score >= 90 else 'B' if avg_score >= 80 else 'C'
    total_amount = sum(r['amount'] for r in recs)
    print(f"{name:<12} {avg_score:>8.1f} {grade:>8} {total_amount:>9}₴")
    summary.append({'name': name, 'avg_score': round(avg_score, 1),
                   'grade': grade, 'total_amount': total_amount})

print()
category_counts = Counter(r['category'] for r in records)
print(f"Витрати по категоріях: {dict(category_counts)}")

# --- Step 4: Збереження (File I/O + JSON) ---
import tempfile, os
report = {
    'generated_at': datetime.now().isoformat(),
    'total_students': len(by_name),
    'students': summary,
    'category_stats': dict(category_counts),
}

fname = tempfile.mktemp(suffix='_report.json')
with open(fname, 'w', encoding='utf-8') as f:
    json.dump(report, f, ensure_ascii=False, indent=2)

print()
print(f"✅ Звіт збережено у {os.path.basename(fname)}")

# Перевіримо що можна прочитати назад
with open(fname, 'r', encoding='utf-8') as f:
    loaded_report = json.load(f)

print(f"✅ Прочитано: {len(loaded_report['students'])} студентів, генеровано {loaded_report['generated_at'][:10]}")
os.unlink(fname)
````

## 10 фундаментальних ідей

### 1. Змінна — ярлик, а не коробка
> `b = a` не копіює значення. Обидва ярлики вказують на один об'єкт.  
> Коли `a` переклеюється — `b` залишається на місці.

### 2. Python динамічний, але строгий
> Тип не треба оголошувати — але Python СУВОРО контролює операції.  
> `"age: " + 25` — TypeError. Явне краще за неявне.

### 3. Структура даних визначає алгоритм
> Потрібен порядок? `list`. Незмінність? `tuple`. Унікальність? `set`.  
> Швидкий пошук? `set` або `dict`. Правильний вибір = простіший код.

### 4. Функція — це рівень абстракції
> Читаючи `apply_tax(bill)` не треба думати про множення.  
> Абстракція — управління складністю.

### 5. Чиста функція — передбачувана функція
> Однакові аргументи → завжди однаковий результат.  
> Не змінює нічого ззовні. Легко тестувати, легко розуміти.

### 6. Помилки — це нормально, якщо їх обробляти
> `try/except` — не милиця, а архітектурне рішення.  
> EAFP > LBYL. Спробуй і обробляй, а не перевіряй кожен сценарій.

### 7. Дані живуть поза програмою
> RAM — тимчасова. Диск — постійний.  
> JSON — мова спілкування між системами.

### 8. Стандартна бібліотека — твій перший ресурс
> Перед `pip install` — перевір stdlib.  
> Python «batteries included» — більшість задач вже вирішена.

### 9. Git — спосіб мислення, а не просто інструмент
> Commit = знімок. Branch = паралельна реальність.  
> PR = спосіб спілкування про зміни.

### 10. Середовище — фундамент всього
> Якщо не знаєш який interpreter запущено — нічого не зрозуміло.  
> venv + правильний kernel = спокійна робота.

> **Програмування — це управління складністю.**  
> Всі патерни, які ми вивчили (DRY, декомпозиція, чисті функції, pipeline),

```python
# Весь Модуль 1 в одному прикладі:

import json
from collections import Counter
from datetime import datetime

def analyze_sales(filename: str) -> dict:        # typing
    """Зчитує продажі з JSON, повертає аналіз."""  # docstring
    try:                                           # exceptions
        with open(filename, encoding='utf-8') as f:  # file I/O
            data = json.load(f)                    # JSON
    except FileNotFoundError:
        return {}

    valid = [r for r in data if r.get('amount', 0) > 0]  # comprehension + dict
    categories = Counter(r['category'] for r in valid)   # counting
    total = sum(r['amount'] for r in valid)               # aggregation

    return {                                              # dict
        'generated': datetime.now().isoformat(),          # datetime
        'total': total,
        'by_category': dict(categories),
        'top': max(categories, key=categories.get),       # reduce
    }

if __name__ == '__main__':                        # __name__
    result = analyze_sales('sales.json')
    print(json.dumps(result, ensure_ascii=False, indent=2))
```
