```python
import pandas as pd

df = pd.read_csv("orders.csv")
df.head(10)
```

```python
total_orders = len(df)
print(total_orders)
total_orders_price = float(df["price"].sum())
meane_price = total_orders_price / total_orders
print(float(meane_price))
print(total_orders_price)
```

```python
report = {
    "total_orders": total_orders,
    "total_orders_price": total_orders_price,
    "meane_price": meane_price,

}
print(report)
```

```python
import json
with open("restaurant_info.json", "w", encoding="utf-8") as file:
    json.dump(report, file, ensure_ascii=False, indent=4)

```

```python
with open("restaurant_info.json", "r", encoding="utf-8") as file:
    saved_report = json.load(file)
    print(saved_report)
    print(saved_report["total_orders"])
    print(saved_report["total_orders_price"])
    print(saved_report["meane_price"])
```
